```python
print("\n Encabezado del dataset")
print(df.head())
```

Informacion general del dataset

```python
print("\n Informacion del dataset")
print(df.info())
print("\n Estadisticas del dataset")
print(df.describe())
```

Conteo de categorias

```python
print("\n Conteo de categorias")
print(df['Categoria'].value_counts())
print("\n Conteo de metodo de pago")
print(df['MetodoPago'].value_counts())
print("\n producto mas caro")
print(df.loc[df["Precio"].idxmax()])
```

```python
print("\n promedio y suma")
print(df['Precio'].mean())
print(df['Cantidad'].sum())
```

```python
df['Categoria'].value_counts().plot(kind='bar')
```

```python
df['Categoria'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Porcentaje de Categorias")
plt.ylabel("")
plt.show()
```

```python
df['Precio'].plot(kind='hist')
plt.title("Distribucion de Precios")
plt.xlabel("")
plt.show()
```

```python
df.plot(kind='scatter',x='Precio',y='Cantidad')
plt.title("Precio vs Cantidad")
plt.show()
```
